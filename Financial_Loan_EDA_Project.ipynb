# Financial Loan + Text Data EDA Project
Complete Exploratory Data Analysis with Business Insights

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import warnings
warnings.filterwarnings('ignore')
import re
from collections import Counter
```

```python
# Load Data
df = pd.read_excel('financial_loan.xlsx')

with open('python financial loan project.txt', 'r', encoding='utf-8') as f:
    text_data = f.read()

df.head()
```

```python
# Basic Info
df.shape
df.info()
df.describe()
df.isnull().sum()
```

```python
# Data Cleaning
df['issue_date'] = pd.to_datetime(df['issue_date'], errors='coerce')
df.drop_duplicates(inplace=True)
df.fillna(method='ffill', inplace=True)
```

```python
# Loan Distribution
sns.histplot(df['loan_amount'], bins=30)
plt.show()

df['loan_status'].value_counts().plot(kind='bar')
plt.show()
```

```python
# Scatter + Box
sns.scatterplot(x='loan_amount', y='int_rate', data=df)
plt.show()

sns.boxplot(x='term', y='loan_amount', data=df)
plt.show()
```

```python
# Time Series
df['month'] = df['issue_date'].dt.to_period('M')
monthly = df.groupby('month')['loan_amount'].sum()
monthly.plot()
plt.show()
```

```python
# KPIs
total_apps = df['id'].count()
total_funded = df['loan_amount'].sum()
total_received = df['total_payment'].sum()

print(total_apps, total_funded, total_received)
```

```python
# Good vs Bad Loans
good = df[df['loan_status'].isin(['Fully Paid','Current'])]
bad = df[df['loan_status']=='Charged Off']

print(len(good)/len(df)*100)
print(len(bad)/len(df)*100)
```

```python
# State Analysis
df.groupby('address_state')['loan_amount'].sum().sort_values(ascending=False).head(10)
```

```python
# Correlation
sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
plt.show()
```

```python
# Text Analysis
words = text_data.split()
print(len(words))

clean_text = re.sub(r'[^a-zA-Z ]','', text_data).lower()
clean_words = clean_text.split()

common = Counter(clean_words).most_common(10)
print(common)
```
