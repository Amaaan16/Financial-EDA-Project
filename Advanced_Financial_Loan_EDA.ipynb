# 📊 Advanced Financial Loan EDA (Question-Based Analysis)
This notebook contains detailed EDA with questions and answers.

```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

## Q1: Load Dataset

```python

df = pd.read_excel('financial_loan.xlsx')
df.head()
```

## Q2: What is dataset shape?

```python
df.shape
```

## Q3: What are data types?

```python
df.info()
```

## Q4: Missing values?

```python
df.isnull().sum()
```

## Q5: Data Cleaning

```python

df['issue_date'] = pd.to_datetime(df['issue_date'], errors='coerce')
df.fillna(method='ffill', inplace=True)
df.drop_duplicates(inplace=True)
```

## Q6: Loan Amount Distribution

```python

sns.histplot(df['loan_amount'], bins=40)
plt.show()
```

## Q7: Loan Status Distribution

```python

df['loan_status'].value_counts().plot(kind='bar')
plt.show()
```

## Q8: Avg Loan per Status

```python

df.groupby('loan_status')['loan_amount'].mean()
```

## Q9: Does interest affect loan amount?

```python

sns.scatterplot(x='loan_amount', y='int_rate', data=df)
plt.show()
```

## Q10: Loan amount by term

```python

sns.boxplot(x='term', y='loan_amount', data=df)
plt.show()
```

## Q11: Monthly trend

```python

df['month'] = df['issue_date'].dt.to_period('M')
df.groupby('month')['loan_amount'].sum().plot()
plt.show()
```

## Q12: Total Applications

```python
df['id'].count()
```

## Q13: Total Funded

```python
df['loan_amount'].sum()
```

## Q14: Total Received

```python
df['total_payment'].sum()
```

## Q15: Good vs Bad Loan %

```python

good = df[df['loan_status'].isin(['Fully Paid','Current'])]
bad = df[df['loan_status']=='Charged Off']

len(good)/len(df)*100, len(bad)/len(df)*100
```

## Q16: Top States

```python

df.groupby('address_state')['loan_amount'].sum().sort_values(ascending=False).head(10)
```

## Q17: Home ownership impact

```python

df.groupby('home_ownership')['loan_amount'].mean()
```

## Q18: Correlation

```python

sns.heatmap(df.corr(), annot=True)
plt.show()
```

## Q19: DTI vs Loan Status

```python

df.groupby('loan_status')['dti'].mean()
```

## Q20: Risk Analysis

```python

df.groupby('loan_status')['int_rate'].mean()
```
